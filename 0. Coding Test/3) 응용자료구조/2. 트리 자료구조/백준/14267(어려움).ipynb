```python
'''

    ,               . ,           .

      ,       .

     ,     ,     ,


     n,    m .  1 n   . (2  n, m  100,000)

   n    .      ,  1 . 1 ,   -1 .

 m       i,   w . (2  i  n, 1  w  1,000)

     .


1 n     .

  1 
5 3
-1 1 2 3 4
2 2
3 4
5 6

  1 
0 2 6 6 12
'''
```

```python
n, m = map(int, input().split())
parent = list(map(int, input().split()))

for i in range(1, n):
    parent[i] -= 1

print('*parent: ', *parent)

good = [0] * n #  
for _ in range(m):
    i, w = map(int, input().split())
    good[i-1] += w #  2, 5 / 2, 3        !!

total_good = [0] * n #   
for i in range(1, n):
    
    print('total_good[' + str(i) + '] = total_good[parent[' + str(i) + ']] + good[' + str(i) + ']')
    print('parent[' + str(i) + ']=' + str(parent[i]))
    print('-------------------------------')
    
    total_good[i] = total_good[parent[i]] + good[i] #  good[parent[i]] + good[i]   ..      

print(*total_good) # print(*arr)      
```
