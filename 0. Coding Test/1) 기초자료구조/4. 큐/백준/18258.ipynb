```python
'''
정수를 저장하는 큐를 구현한 다음, 입력으로 주어지는 명령을 처리하는 프로그램을 작성하시오.

명령은 총 여섯 가지이다.

push X: 정수 X를 큐에 넣는 연산이다.
pop: 큐에서 가장 앞에 있는 정수를 빼고, 그 수를 출력한다. 만약 큐에 들어있는 정수가 없는 경우에는 -1을 출력한다.
size: 큐에 들어있는 정수의 개수를 출력한다.
empty: 큐가 비어있으면 1, 아니면 0을 출력한다.
front: 큐의 가장 앞에 있는 정수를 출력한다. 만약 큐에 들어있는 정수가 없는 경우에는 -1을 출력한다.
back: 큐의 가장 뒤에 있는 정수를 출력한다. 만약 큐에 들어있는 정수가 없는 경우에는 -1을 출력한다.
'''

import sys

N = int(input()) # 여기 int로 감싸지 않아서 
queue = []
for _ in range(N): # 여기서 에러남....감싸지않으면 문자열이기 때문에..
    # i = input() 이거 작동은 하는데 시간 초과 발생... 그래서 아래와 같이 고쳐줌
    i = sys.stdin.readline().rstrip() # rstrip은 \n(개행) 제거
    
    if 'push' in i:
        num = int(i[-1])
        queue.append(num)
    elif i == 'pop':
        if not queue:
            print(-1)
        else:
            print(queue.pop(0))
    elif i == 'size':
        print(len(queue))
    elif i == 'empty':
        if not queue:
            print(1)
        else:
            print(0)
    elif i == 'front':
        if not queue:
            print(-1)
        else:
            print(queue[0])
    elif i == 'back':
        if not queue:
            print(-1)
        else:
            print(queue[len(queue)-1]) # 이렇게 할 필요 없고 그냥 queue[-1]로 하면 돼 


'''
실행하면 input값 입력하는게 안나와서.. 도대체 어떻게 해야하는건지 모르겠음...
글고 이렇게 백준에 제출하면 에러남..(틀렸다고 나옴..)
'''
```

```python
# 선생님 풀이

import sys
from collections import deque #데크라고 읽음~~ 큐 자료구조 사용할때 사용 ⭐

N = int(input())
d = deque([]) #데크 선언!!!(외우기) ⭐

for _ in range(N):
    command = sys.stdin.readline() # input()보다 입력을 빠르게 받을수 있음.(많은 입력을 받아야할 때 사용)
    command = command.split() # 매개변수에 아무것도 입력하지 않으면 -> 공백 기준으로 split. 여러 공백은 하나의 공백으로 처리.(+ 앞뒤 공백 제거)

    if command[0] == 'push':
        d.append(command[1]) # 숫자로 변환 안하고 문자열 그 자체로 append해도 괜찮다. 이 문제에선. 왜냐면 연산을 할 게 아니기 때문에..
    if command[0] == 'pop':
        if not d:
            print(-1)
        else:
            print(d.popleft()) # 여기 !!!!!!!!!!!!!!!!! ⭐
    if command[0] == 'size':
        print(len(d))
    if command[0] == 'empty':
        if not d:
            print(1)
        else:
            print(0)
    if command[0] == 'front':
        if not d:
            print(-1)
        else:
            print(d[0])
    if command[0] == 'back':
        if not d:
            print(-1)
        else:
            print(d[-1])
```
