3052 - 나머지

문제
두 자연수 A와 B가 있을 때, A%B는 A를 B로 나눈 나머지 이다. 예를 들어, 7, 14, 27, 38을 3으로 나눈 나머지는 1, 2, 0, 2이다. 

수 10개를 입력받은 뒤, 이를 42로 나눈 나머지를 구한다. 그 다음 서로 다른 값이 몇 개 있는지 출력하는 프로그램을 작성하시오.

입력
첫째 줄부터 열번째 줄 까지 숫자가 한 줄에 하나씩 주어진다. 이 숫자는 1,000보다 작거나 같고, 음이 아닌 정수이다.

출력
첫째 줄에, 42로 나누었을 때, 서로 다른 나머지가 몇 개 있는지 출력한다.

```python
r_set = set() #빈 set 선언할떄 이렇게 하기

count = 0

while True:
    if count >= 10: # ⚠️ 이거 count > 10 이 아니라 count >= 10임!!(완전 중요. 여기서 틀림 ㅠㅠ). 아니면 아예 count == 10으로 하면 될듯!!
        #왜냐면 마지막에 10번째에서 set에 추가되고, 다시 while True의 처음으로 들어올땐 일단 count가 10이기 때문에 >10라면 안으로 들어올수있음
        break
        
    input_value = int(input())
    if input_value > 0 and input_value <= 1000:
        count += 1 # ⚠️ 파이썬엔 증감연산자가 없음!!! (count++ 로 썼다가 에러남)
        r_set.add(input_value%42) # ⚠️ r_set을 r.set으로 오타나서 에러남 # 그리고 참고적으로 set은 append가 아니라 add임!!(순서가 없기 때문에)
        
print(len(r_set))
```

================================================================= <br>
나는 위에서 set으로 풀었지만 이거를 체크배열로 풀 수 있다.(선생님 풀이)

나머지가 배열의 인덱스가 되고, 그 나머지가 계산에서 나왔느냐의 여부(나왔으면 1, 안나왔으면 0)가 그 값..

예를 들어, 나머지가(값이) 5이면 그 부분(check_array[5]을 0에서 1로 바꿔줌(0->1). (1이었으면 그대로 둠)(1->1) <br>
그래서 1인 애들의 총 갯수(배열의 총합)(0+0+1+0+1...)을 구하면 .. 그게 이 문제의 답이다. 




```python
check = [0] * 42

for _ in range(10):
    n = int(input())
    check[n % 42] = 1

sum = 0
for c in check:
    sum += c

'''
또는
for i in range(42):
    sum += check[i]
'''

print(sum)
```
